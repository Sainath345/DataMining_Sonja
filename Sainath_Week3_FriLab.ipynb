```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

#Reading in the data as ipl dataframe
ball = pd.read_csv("https://raw.githubusercontent.com/Sainath345/Rasaputra_IT533/main/Ball_by_Ball.csv")

#Verifying that we can see the data
ball.head()
```

```python
ball_2 = ball[ball['Extra_Type']!= ' ']
ball_2_small = ball_2.iloc[:100]
ball_2_small.Extra_Type.unique()
```

```python
# Creating a new attribute and populating it with the contents of the attribute that we want to transform.
Extra_type_code =ball_2_small.Extra_Type

# Adding the new attribute to the dataframe
ball_2_small['Extra_type_code'] = Extra_type_code

# Checking the contents of the new attribute
ball_2_small
```

```python
ball_2_small['Extra_type_code'] = ball_2_small['Extra_type_code'].astype('category')
ball_2_small['Extra_type_code'] = ball_2_small['Extra_type_code'].cat.codes
ball_2_small['Extra_type_code']
```
